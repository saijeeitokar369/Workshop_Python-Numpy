# Broadcasting

```python
k= np.arange(6).reshape(2,3)
print(k)
print()

l=np.array([100,200,300])
print(l)
sum=k+l
print()
print(sum)
```

```python
k= np.arange(6).reshape(2,3)
print(k)
print()

l=np.array([100,200,300])
print(l)
sum=k-l
print()
print(sum)
```
